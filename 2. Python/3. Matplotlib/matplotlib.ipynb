## Multiplots

### Method 1:using the add_axes() method

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

fig=plt.figure(figsize=(5,4))

ax1=fig.add_axes([0.1,0.1,0.8,0.8]) 
ax2=fig.add_axes([0.5,0.5,0.3,0.3]) 

ax1.plot([5,4,3,2,1],[2,3,4,5,6])
ax2.plot([1,2,3,4,5],[2,3,4,5,6])

plt.show()
```

### Method 2: using subplot() method

```python
x=[3,1,3]
y=[3,2,1]
z=[1,3,1]
# fig object
plt.figure()

#1st plot
plt.subplot(121) #1st row,2nd column,1st plot
plt.plot(x,y)

#2nd plot
plt.subplot(122) #1st row,2nd column,2nd plot
plt.plot(z,y)
```

### Method 2: using subplots() method

### tick labels

```python
x=[3,1,3]
y=[3,2,1]

fig=plt.figure(figsize=(5,4))

ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
ax.set_xlim(1,2)
ax.set_xticklabels(('one','two','three','four','five','six'))

plt.show()
```

## Annotations

```python
fig,ax=plt.subplots(figsize=(5,2.7))
t=np.arange(0.0,5.0,0.01)
s=np.cos(2*np.pi*t)
line=ax.plot(t,s,lw=2)
ax.annotate('local max',xy=(2,1),xytext=(3,1.5),
           arrowprops=dict(facecolor='black',shrink=0.01))
ax.set_ylim(-2,2);
```

## pie chart

```python
players='virat','rohit','shikhar','hardik'
runs=[31,52,71,53]
explode=(0,0,0.1,0)
fig1,ax1=plt.subplots()
ax1.pie(runs,explode=explode,labels=players,autopct='%1.3f%%',
       shadow=True, startangle=180)
ax1.axis('equal')
```

## Area plot or stack plot

```python
days=[1,2,3,4,5]

sleeping=[4,5,6,7,8]
eating=[7,6,4,3,2]
working=[9,3,6,4,7]
playing=[7,3,2,5,9]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.xlabel('y')
plt.title('Stack plot')
plt.legend()
plt.show()
```
